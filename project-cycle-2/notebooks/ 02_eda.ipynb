```python
sns.histplot(df['BMIPCT'], kde=False)
plt.title("BMI Percentile")
plt.show()
```

The histogram shows the distribution of BMI percentiles.The distribution is not symmetric and appears to be skewed to the left, as more observations are concentrated at higher BMI percentiles.Most students have BMI percentiles between approximately 60 and 100, with a noticeable concentration near the upper end.This suggests that a large proportion of students have relatively high BMI percentiles.

```python
bmi = df['BMIPCT'].dropna()

sns.boxplot(x=bmi)

plt.title("Boxplot of BMI Percentile")
plt.xlabel("BMIPCT")

plt.show()
```

The boxplot shows the distribution of BMI percentiles.The median is located around the center of the box, indicating the central tendency of the data.The spread between the first and third quartiles suggests the variability of the middle 50% of observations.Some potential outliers can be observed outside the whiskers.

```python
Q1 = df['BMIPCT'].quantile(0.25)
Q3 = df['BMIPCT'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['BMIPCT'] < Q1 - 1.5*IQR) | (df['BMIPCT'] > Q3 + 1.5*IQR)]

print("Number of outliers:", len(outliers))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 先處理資料
extra_df = df[df["EverCigaretteUse"].isin([1, 2])].copy()

# 轉成文字（比較好看）
extra_df["smoke_group"] = extra_df["EverCigaretteUse"].map({
    1: "Ever smoked",
    2: "Never smoked"
})

# 處理 BMI
extra_df["BMIPCT"] = pd.to_numeric(extra_df["BMIPCT"], errors="coerce")
extra_df = extra_df.dropna(subset=["BMIPCT"])

# 畫 boxplot
sns.boxplot(x="smoke_group", y="BMIPCT", data=extra_df)
plt.title("BMI Percentile by Smoking Status")
plt.xlabel("Smoking Status")
plt.ylabel("BMI Percentile")
plt.show()
```

This plot shows how BMI is distributed among students who have smoked. Most values are around 70, but the range is wide, meaning BMI varies a lot.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 複製資料
gender_df = df.copy()

# 看 sex 原始編碼（先確認）
print(gender_df["WhatIsYourSex"].value_counts(dropna=False))

# 👉 常見情況：1 = male, 2 = female（如果不是再改）
gender_df = gender_df[gender_df["WhatIsYourSex"].isin([1, 2])]

gender_df["gender"] = gender_df["WhatIsYourSex"].map({
    1: "Male",
    2: "Female"
})

# BMI 處理
gender_df["BMIPCT"] = pd.to_numeric(gender_df["BMIPCT"], errors="coerce")
gender_df = gender_df.dropna(subset=["BMIPCT"])

# 檢查兩組都有
print(gender_df["gender"].value_counts())

# 畫圖
sns.boxplot(x="gender", y="BMIPCT", data=gender_df)
plt.title("BMI Percentile by Gender")
plt.xlabel("Gender")
plt.ylabel("BMI Percentile")
plt.show()
```

Female BMI percentile median is slightly higher than males, but distributions overlap greatly, indicating no significant overall difference
