EverCigaretteUse Inference

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv("YRBS_2007.csv", encoding="latin1")
pd.set_option('display.max_columns', None)
df.head()
```

```python
# 建立 binary（1=有抽過, 2=沒有）
df['cig_binary'] = df['EverCigaretteUse'].apply(lambda x: 1 if x == 1 else 0)

p_hat = df['cig_binary'].mean()
n = df['cig_binary'].count()

print("Sample proportion:", p_hat)
print("Sample size:", n)
```

```python
p0 = 0.50
print("Benchmark proportion (p0) =", p0)
```

```python
import numpy as np

se = np.sqrt(p_hat * (1 - p_hat) / n)

lower = p_hat - 1.96 * se
upper = p_hat + 1.96 * se

print("95% CI:", (lower, upper))
```

```python
from scipy import stats

z = (p_hat - p0) / se
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Z statistic:", z)
print("p-value:", p_value)
```

The sample proportion of students who have ever used cigarettes is 0.5102200697955986.
The 95% confidence interval is  (0.501951387355792, 0.5184887522354052).
The p-value is 0.01541179870653342.
Since p-value < 0.05, we reject the null hypothesis.（or ≥ 0.05 → fail to reject）
This suggests that the true proportion is different from  0.50.

```python
mean = df['BMIPCT'].mean()
std = df['BMIPCT'].std()
n = df['BMIPCT'].count()

print("Mean:", mean)
print("Std:", std)
print("n:", n)
```

```python
mu0 = 65
print("Benchmark mean μ₀:", mu0)
```

```python
se = std / np.sqrt(n)

lower = mean - 1.96 * se
upper = mean + 1.96 * se

print("95% CI:", (lower, upper))
```

```python
from scipy import stats

t_stat, p_value = stats.ttest_1samp(df['BMIPCT'].dropna(), 65)

print("t-stat:", t_stat)
print("p-value:", p_value)
```

The sample mean BMI percentile is 64.82068317411306,standard deviation is 27.516756256180642 and sanple size is 13062.
The 95% confidence interval is (64.3487848135942, 65.29258153463192).
The p-value is 0.4564174644584005.
Since p-value > 0.05, we don't reject the null hypothesis.
This suggests that the mean BMI percentile is not different from  65.

Interpretation EverCigaretteUse

For the proportion analysis, we estimated the population proportion of students who have ever smoked.
We also tested whether this population proportion is different from the benchmark value of 0.50.

Because the p-value is greater than or equal to 0.05, we fail to reject the null hypothesis.
This means that there is not enough evidence to conclude that the population proportion of students who have ever smoked is different from 0.50.

This result is consistent with the EDA, which showed that the proportion of students who had ever smoked appeared to be relatively high.

The results should be interpreted cautiously because missing values, possible outliers, and the non-symmetric distribution of the data may affect the analysis.
In addition, statistical significance does not necessarily imply practical importance.
