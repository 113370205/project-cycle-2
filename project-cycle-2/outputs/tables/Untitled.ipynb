```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv("yrbs_cleaned.csv", encoding="latin1")
```

```python
import pandas as pd

# proportion
n_prop = len(df)
p_hat = df["EverCigaretteUse"].mean()

# mean
n_mean = len(df)
mean_bmi = df["BMIPCT"].mean()
std_bmi = df["BMIPCT"].std()

# 建表
eda_summary = pd.DataFrame({
    "Variable": ["EverCigaretteUse", "BMIPCT"],
    "Type": ["Binary", "Continuous"],
    "Sample Size": [n_prop, n_mean],
    "Mean / Proportion": [p_hat, mean_bmi],
    "Standard Deviation": ["-", std_bmi],
})

eda_summary
eda_summary.to_csv("eda_summary_table.csv", index=False)
```

```python
import pandas as pd

inference_summary = pd.DataFrame({
    "Analysis": ["Proportion", "Mean"],
    "Parameter": ["p", "μ"],
    "Estimate": [0.5102, 64.82],
    "95% CI": [
        "(0.50195, 0.51849)",
        "(64.35, 65.29)"
    ],
    "p-value": [0.0154, 0.4564],
    "Decision": ["Reject H0", "Fail to reject H0"]
})

# 存成 CSV
inference_summary.to_csv("inference_summary_table.csv", index=False)

inference_summary
```
